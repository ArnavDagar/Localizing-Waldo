```python
# Generate color strips
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import random
import matplotlib.image as mpimg
import cv2 as cv

'''
COLOR_MAP = {
    Color.NOCOLOR: 0,
    Color.BLACK: 1,
    Color.BLUE: 2,
    Color.GREEN: 3,
    Color.YELLOW: 4,
    Color.RED: 5,
    Color.WHITE: 6,
    Color.BROWN: 7
}
'''
'''Color Code From PyBricks and other sources '''
colorCode = np.zeros((8,3), np.uint8)
colorCode[0] = [0, 0, 0]
colorCode[1] = [26, 26, 26]
colorCode[2] = [0, 0, 255]
colorCode[3] = [0, 150, 57]
colorCode[4] = [255, 255, 0]
colorCode[5] = [255, 0, 0]
colorCode[6] = [255, 255, 255]
colorCode[7] = [96, 61, 32]

mmPerInch = 25.4
dpi = 200 # Printer dpi set to 200
pixelsPerMm = dpi / mmPerInch

def color_transition(id):
    color_ids = np.arange(1,8)
    markers = np.full(7,id)
    c = np.column_stack((markers, color_ids)).flatten()
    c1 = np.append(c[1:],0)
    d = c[c!=c1]
    if id == d[-1]:
        return d
    return np.append(d,id)

def generate_color_strips(id1, id2):
    # Each strop will have 13 24 mm x 50 mm colored rectangles
    # Each page can have 2 strips with a white strip in the middle 
    nR = 3
    nC = 13
    world = np.zeros((nR,nC),np.uint8)
    world[0] = color_transition(id1)
    world[2] = color_transition(id2)
    pixelsC = round(24*pixelsPerMm)
    pixelsR = round(50*pixelsPerMm)
    img = np.zeros((pixelsR*nR,pixelsC*nC,3),np.uint8)
    for i in range(nR):
        for j in range(nC):
            for k in range(pixelsR):
                for l in range(pixelsC):
                    if i == 0:
                        img[pixelsR*i+k, pixelsC*j+l,:] = colorCode[world[i,j]]
                    elif i == 1:
                        img[pixelsR*i+k, pixelsC*j+l,:] = [255, 255, 255]
                    else:
                        img[pixelsR*i+k, pixelsC*j+l,:] = colorCode[world[i,j]]
    return img

f,axes = plt.subplots(1,3,figsize=(15,15))
f.tight_layout()

img = generate_color_strips(5,7)
axes[0].imshow(img)
fileName = 'Strip 5, 7.jpg'
plt.imsave(fileName, img, dpi = 200)

img = generate_color_strips(3,4)
axes[1].imshow(img)
fileName = 'Strip 3, 4.jpg'
plt.imsave(fileName, img, dpi = 200)

img = generate_color_strips(1,2)
axes[2].imshow(img)
fileName = 'Strip 1, 2.jpg'
plt.imsave(fileName, img, dpi = 200)
    
```
